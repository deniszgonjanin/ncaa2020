```python
import pandas as pd
```

```python
# One row per matchup per season
# | Season | Team_ID-L | Team_ID-H | Outcome (1,0)
```

```python
validation_df = pd.read_csv("data/2020-Mens-Data/MDataFiles_Stage1/MNCAATourneyCompactResults.csv")
```

```python
validation_df.head()
```

```python
validation_df = validation_df.drop(["DayNum", "WLoc", "LScore", "WScore", "NumOT"], axis=1)
```

```python
validation_df['Team_ID_L'] = validation_df[['WTeamID', 'LTeamID']].min(axis=1)
validation_df['Team_ID_H'] = validation_df[['WTeamID', 'LTeamID']].max(axis=1)
```

```python
def assign_winner_low_or_high(row):
    if row['WTeamID'] == row['Team_ID_L']:
        return 1
    else:
        return 0


validation_df['Outcome'] = validation_df.apply(lambda row: assign_winner_low_or_high(row), axis=1)
```

```python
validation_df[0:20]
```

```python
validation_df = validation_df.drop(["WTeamID", "LTeamID"], axis=1)
```
